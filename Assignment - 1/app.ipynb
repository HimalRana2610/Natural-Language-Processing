```python
from datasets import load_dataset

dataset = load_dataset("ai4bharat/IndicCorpV2", "indiccorp_v2", split="guj_Gujr", streaming=True)
```

```python
import re

def sentence_tokenizer(text):
  text = re.sub(r'\s+', ' ', text.strip())
  sentences = re.split(r'(?<=[.!?])\s+', text)
  return [s.strip() for s in sentences if s.strip()]
```

```python
def word_tokenizer(text):
  token_patterns = [
      r'https?://\S+',                           # URLs
      r'\b[\w\.-]+?@\w+\.\w+\b',                 # Email addresses
      r'\b\d{1,2}[/-]\d{1,2}[/-]\d{2,4}\b',      # Dates
      r'\d+\.\d+|\d+',                           # Numbers
      r'[^\w\s\u0A80-\u0AFF]',                   # Non-Gujarati/English punctuation
      r'[\u0A80-\u0AFF]+',                       # Gujarati words
      r'[A-Za-z]+',                              # English words
  ]
  token_re = re.compile('|'.join(token_patterns))
  return token_re.findall(text)
```

```python
with open("sentences.txt", "w", encoding="utf-8") as f:
  for i, data in enumerate(dataset):
    sentences = sentence_tokenizer(data["text"])
    for s in sentences:
      f.write(s + "\n")
    if i == 100:
      break;
print("Tokenized sentences stored in sentences.txt")
```

```python
with open("sentences.txt", "r", encoding="utf-8") as f:
  sentences = f.readlines()
  for s in sentences:
    words = word_tokenizer(s)
    for word in words:
      print(word + " ", end="")
    print()
```

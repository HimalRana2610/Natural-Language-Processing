```python
from visual_automata.fa.dfa import VisualDFA
import string

dfa = VisualDFA(
    states={"q0", "q1", "q2", "Dead"},
    input_symbols={ch for ch in string.printable},
    transitions={
        "q0": {ch: "q1" for ch in string.ascii_uppercase + string.ascii_lowercase} | {ch: "Dead" for ch in string.printable if ch not in string.ascii_uppercase + string.ascii_lowercase},
        "q1": {ch: "q2" for ch in string.ascii_lowercase} | {ch: "Dead" for ch in string.printable if ch not in string.ascii_lowercase},
        "q2": {ch: "q1" for ch in string.ascii_lowercase} | {ch: "Dead" for ch in string.printable if ch not in string.ascii_lowercase},
        "Dead": {ch: "Dead" for ch in string.printable}
    },
    initial_state="q0",
    final_states={"q1", "q2"},
)
```

```python
class MorphFST:
    def __init__(self):
        self.START = "q0"
        self.IES = "q_ies"
        self.ES = "q_es"
        self.SIMPLE_S = "q_s"
        self.ACCEPT_SG = "q_accept_sg"
        self.ACCEPT_PL = "q_accept_pl"
        self.REJECT = "q_reject"

    def process(self, word: str) -> str:
        state = self.START
        root = ""
        w = word.lower()

        special_endings = ("s", "z", "x", "ch", "sh")

        if not w.endswith("s"):
            return f"{w}+N+SG"

        if w.endswith("ies"):
            stem = w[:-3]
            if stem and stem[-1] not in "aeiou":
                return f"{stem}y+N+PL"
            else:
                return "Invalid Word"

        if w.endswith("es"):
            stem = w[:-2]
            if stem.endswith(special_endings):
                return f"{stem}+N+PL"
            else:
                return "Invalid Word"

        if w.endswith("s"):
            stem = w[:-1]
            if not stem.endswith(special_endings):
                return f"{stem}+N+PL"
            else:
                return "Invalid Word"

        return "Invalid Word"
```

```python
fst = MorphFST()

with open("brown_nouns.txt", "r", encoding="utf-8") as in_f:
  nouns = [line.strip() for line in in_f if line.strip()]

with open("output.txt", "w", encoding="utf-8") as out_f:
  for word in nouns:
    out_f.write(f"{fst.process(word)}\n")

print("Results saved in output.txt")
```
